```python
import pandas as pd
import os
import gensim
import spacy
```

```python
#Loading the NLP Model
nlp = spacy.load('en_core_web_lg')
```

```python
#Loading Training data
clean_data = pd.read_csv('train_data.csv')
```

```python
clean_data.rename(columns = {'_col2':'experience'}, inplace = True)
clean_data.head(5)
```

```python
# Data Profile
clean_data.info()
```

```python
#converting the training dataset to all strings as it my have any other data type in a row
clean_data['title'] = clean_data['title'].astype(str)
```

```python
#training sentences Tokenized
train_tokens = [nlp(row) for row in clean_data['title']]
```

```python
train_tokens
```

```python
type(train_tokens)
```

```python
#test sentence on which the model will be tested
test_job = 'security'
test_exp = 12
test_city = 'lahore'
```

```python
test_job = test_job.title()
test_city = test_city.title()
```

```python
#initaillising to get results in the lists, index shows the row number of the similar sentance, sim_score shows the similarity score
sim_score = ['sim_score']
index = ['index']
```

```python
#test sentence Tokenized
test_token = nlp(test_job)
```

```python
# looping through each training set and checking for simlarity with the test_sentence (token)
# Svaing the results in the sim_score list along with the index of the result in idex list
# When the maps get 5 Job Titles with the test Input (test_token) it exits the loop to save time
for i in range(len(train_tokens)):
    sim = train_tokens[i].similarity(test_token)
    if sim >= 0.85 and sim <= 1:
        if test_exp >= clean_data['experience'].iloc[i] and test_city == clean_data['city'].iloc[i]:
            sim_score.append(sim)
            index.append(clean_data['jid'][i])
            sim_score_doc = pd.DataFrame(list(zip(index, sim_score)), columns = ['jid', 'sim_score'])
    elif len(index) == 21:
        break
```
